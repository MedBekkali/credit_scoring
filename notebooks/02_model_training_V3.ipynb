```python
# 02 - Entraînement des modèles (LogReg + LightGBM)
import sys
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import mlflow
import mlflow.sklearn

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import (
    StratifiedKFold,
    cross_val_score,
    train_test_split,
)
from sklearn.pipeline import Pipeline

from lightgbm import LGBMClassifier

# Import des fonctions du projet
sys.path.append("..")
from src.data_prep import build_datasets, build_preprocessor
from src.metrics import compute_classic_metrics, business_cost, cost_curve

# Configuration MLflow (nouveau dossier mlruns_V_F)


# On pointe vers un nouveau dossier de tracking : mlruns_V_F
PROJECT_ROOT = Path("..").resolve()
MLFLOW_DIR = PROJECT_ROOT / "mlruns_V_F"

mlflow.set_tracking_uri(f"file:{MLFLOW_DIR}")
mlflow.set_experiment("credit_scoring")

mlflow.get_tracking_uri()

# 1. Chargement des données et préprocesseur


train_df, test_df = build_datasets()

X = train_df.drop(columns=["TARGET"])
y = train_df["TARGET"]

preprocessor = build_preprocessor(train_df)

print("Shapes train / test :", train_df.shape, test_df.shape)
print("Taux de défaut (TARGET=1) :", y.mean())

# 2. Déséquilibre des classes


n_pos = int(y.sum())
n_neg = len(y) - n_pos
imbalance_ratio = n_neg / n_pos

print(f"Nombre de bons payeurs (0) : {n_neg}")
print(f"Nombre de mauvais payeurs (1) : {n_pos}")
print(f"Ratio négatifs / positifs : {imbalance_ratio:.2f}")

# 3. Modèle baseline : régression logistique (avec gestion du déséquilibre)


log_reg = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",  # corrige le déséquilibre
    n_jobs=-1,
)

log_reg_pipeline = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", log_reg),
])

# Cross-validation stratifiée (3 folds pour aller plus vite)
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

# On peut travailler sur un sous-échantillon pour éviter des temps énormes
X_small, _, y_small, _ = train_test_split(
    X,
    y,
    test_size=0.7,  # on garde 30 % pour la CV baseline
    stratify=y,
    random_state=42,
)

log_scores = cross_val_score(
    log_reg_pipeline,
    X_small,
    y_small,
    cv=cv,
    scoring="roc_auc",
    n_jobs=1,  # pas de parallélisation pour éviter les soucis Windows
)

print(f"AUC moyenne (LogReg) : {log_scores.mean():.4f}")
print(f"Écart-type AUC       : {log_scores.std():.4f}")

# Tracking MLflow pour le baseline
with mlflow.start_run(run_name="logreg_baseline"):
    mlflow.log_param("model_type", "logistic_regression")
    mlflow.log_param("max_iter", log_reg.max_iter)
    mlflow.log_param("class_weight", "balanced")

    mlflow.log_metric("cv_auc_mean", float(log_scores.mean()))
    mlflow.log_metric("cv_auc_std", float(log_scores.std()))

    # Entraînement sur le sous-échantillon et log du modèle
    log_reg_pipeline.fit(X_small, y_small)
    mlflow.sklearn.log_model(log_reg_pipeline, "model")

# 4. Modèle avancé : LightGBM (boosting de gradient)


lgbm = LGBMClassifier(
    n_estimators=300,
    learning_rate=0.05,
    num_leaves=31,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="binary",
    scale_pos_weight=imbalance_ratio,  # pondération FN/FP via déséquilibre
    random_state=42,
    n_jobs=-1,
)

lgbm_pipeline = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", lgbm),
])

# Cross-validation sur un échantillon (plus gros que pour la LogReg si tu veux)
X_small_lgbm, _, y_small_lgbm, _ = train_test_split(
    X,
    y,
    test_size=0.5,  # 50 % des données
    stratify=y,
    random_state=42,
)

lgbm_scores = cross_val_score(
    lgbm_pipeline,
    X_small_lgbm,
    y_small_lgbm,
    cv=cv,
    scoring="roc_auc",
    n_jobs=1,
)

print(f"AUC moyenne (LightGBM) : {lgbm_scores.mean():.4f}")
print(f"Écart-type AUC         : {lgbm_scores.std():.4f}")

with mlflow.start_run(run_name="lgbm_baseline"):
    mlflow.log_param("model_type", "lightgbm")
    mlflow.log_param("n_estimators", lgbm.n_estimators)
    mlflow.log_param("learning_rate", lgbm.learning_rate)
    mlflow.log_param("num_leaves", lgbm.num_leaves)
    mlflow.log_param("subsample", lgbm.subsample)
    mlflow.log_param("colsample_bytree", lgbm.colsample_bytree)
    mlflow.log_param("scale_pos_weight", imbalance_ratio)

    mlflow.log_metric("cv_auc_mean", float(lgbm_scores.mean()))
    mlflow.log_metric("cv_auc_std", float(lgbm_scores.std()))

    # Entraînement sur l'échantillon et log modèle
    lgbm_pipeline.fit(X_small_lgbm, y_small_lgbm)
    mlflow.sklearn.log_model(lgbm_pipeline, "model")

# 5. Coût métier et optimisation du seuil (FN = 10 × FP)


# Split train / validation sur tout le dataset pour le choix du seuil
X_train, X_valid, y_train, y_valid = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42,
)

# Nouveau pipeline LightGBM entraîné sur X_train
lgbm_pipeline = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", lgbm),
])

lgbm_pipeline.fit(X_train, y_train)

# Probabilités de défaut sur le jeu de validation
y_valid_proba = lgbm_pipeline.predict_proba(X_valid)[:, 1]

# Courbe coût vs seuil (FN = 10, FP = 1)
thresholds, costs = cost_curve(
    y_valid,
    y_valid_proba,
    cost_fn=10.0,
    cost_fp=1.0,
    normalize=True,
)

best_idx = costs.argmin()
best_threshold = float(thresholds[best_idx])
best_cost = float(costs[best_idx])

print(f"Seuil optimal : {best_threshold:.2f}  —  coût moyen : {best_cost:.4f}")

# 6. Courbe coût métier vs seuil + log dans MLflow


fig_path = "../reports/figures/courbe_cout_vs_seuil.png"

plt.figure()
plt.plot(thresholds, costs, marker="o")
plt.axvline(best_threshold, linestyle="--")
plt.xlabel("Seuil de décision")
plt.ylabel("Coût métier moyen par client")
plt.title("Coût métier en fonction du seuil (FN=10, FP=1)")
plt.tight_layout()
plt.savefig(fig_path, bbox_inches="tight")
plt.show()

with mlflow.start_run(run_name="lgbm_cost_curve"):
    mlflow.log_param("model_type", "lightgbm")
    mlflow.log_param("business_cost_fn", 10.0)
    mlflow.log_param("business_cost_fp", 1.0)
    mlflow.log_metric("best_threshold", best_threshold)
    mlflow.log_metric("best_cost", best_cost)
    mlflow.log_artifact(fig_path, artifact_path="figures")

# 7. Comparaison seuil 0,5 vs seuil optimisé (métriques + coût métier)


metrics_05 = compute_classic_metrics(y_valid, y_valid_proba, threshold=0.5)
cost_05, conf_05 = business_cost(y_valid, y_valid_proba, threshold=0.5)

metrics_opt = compute_classic_metrics(y_valid, y_valid_proba, threshold=best_threshold)
cost_opt, conf_opt = business_cost(y_valid, y_valid_proba, threshold=best_threshold)

print("=== Seuil 0.5 ===")
print(metrics_05)
print(f"Coût moyen : {cost_05:.4f}")
print(f"Confusion matrix : {conf_05}")

print("\n=== Seuil optimisé ===")
print(metrics_opt)
print(f"Coût moyen : {cost_opt:.4f}")
print(f"Confusion matrix : {conf_opt}")

# 8. Entraînement et export du modèle final LightGBM (pour serving)


final_lgbm = LGBMClassifier(
    n_estimators=300,
    learning_rate=0.05,
    num_leaves=31,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="binary",
    scale_pos_weight=imbalance_ratio,
    random_state=42,
    n_jobs=-1,
)

final_pipeline = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", final_lgbm),
])

# Entraînement sur tout le train
final_pipeline.fit(X, y)

FINAL_THRESHOLD = best_threshold

# Cross-validation pour vérifier la robustesse
final_cv_scores = cross_val_score(
    final_pipeline,
    X,
    y,
    cv=cv,
    scoring="roc_auc",
    n_jobs=1,
)

with mlflow.start_run(run_name="lgbm_final_model"):
    mlflow.log_param("model_type", "lightgbm")
    mlflow.log_param("n_estimators", final_lgbm.n_estimators)
    mlflow.log_param("learning_rate", final_lgbm.learning_rate)
    mlflow.log_param("num_leaves", final_lgbm.num_leaves)
    mlflow.log_param("subsample", final_lgbm.subsample)
    mlflow.log_param("colsample_bytree", final_lgbm.colsample_bytree)
    mlflow.log_param("scale_pos_weight", imbalance_ratio)
    mlflow.log_param("business_threshold", FINAL_THRESHOLD)

    mlflow.log_metric("final_auc_mean", float(final_cv_scores.mean()))
    mlflow.log_metric("final_auc_std", float(final_cv_scores.std()))

    y_train_proba = final_pipeline.predict_proba(X)[:, 1]
    train_cost, _ = business_cost(y, y_train_proba, threshold=FINAL_THRESHOLD)
    mlflow.log_metric("train_business_cost", float(train_cost))

    # 1) Log du modèle + enregistrement dans le model registry
    mlflow.sklearn.log_model(
        sk_model=final_pipeline,
        artifact_path="model",
        registered_model_name="credit_scoring_model",
    )

    # 2) Export du modèle dans ../model pour Docker / serving
    mlflow.sklearn.save_model(
        sk_model=final_pipeline,
        path="../model",
    )

print(f"Modèle LightGBM final exporté dans ../model avec seuil {FINAL_THRESHOLD:.2f}")
```
